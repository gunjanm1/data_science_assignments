Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.

* Series can only contain single list with index, whereas dataframe can be made of more than one series or we can say that a dataframe is a collection of series that can be used to analyse the data.
* Series elements must be of the same data type while DataFrame elements can have different data types.
*  Once created, the size of a Series object cannot be changed while elements can be added in data frame.

```python
auth_series = pd.Series(['Jitender', 'Purnima', 'Arpit', 'Jyoti'])
print(auth_series)
article_series = pd.Series([210, 211, 114, 178])
print(article_series)
frame={ 'Author': auth_series, 'Article': article_series }
df3=pd.DataFrame(frame)
df3
```
